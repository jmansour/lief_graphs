```python
%matplotlib notebook  
import numpy as np
import matplotlib.pyplot as plt
```

```python
guys = [
        ( 135. , '(M3/1) inception3' , 'lightblue', '///'),
        ( 212. ,'(DGX/1) inception3' ,   'skyblue', '///'),
        (1531. ,'(DGX/8) inception3' ,   'skyblue', '///'),
        ( 229. ,'(EC2/8) inception3' ,     'khaki', '\\\\\\'),
        (1783. ,'(EC2/64) inception3',     'khaki', '\\\\\\'),

        ( 214. ,    '(M3/1) resnet50', 'lightblue', '///'),
        ( 327. ,   '(DGX/1) resnet50',   'skyblue', '///'),
        (2218. ,   '(DGX/8) resnet50',   'skyblue', '///'),
        ( 378. ,   '(EC2/8) resnet50',     'khaki', '\\\\\\'),
        (2744. ,  '(EC2/64) resnet50',     'khaki', '\\\\\\')
       ]
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
# ax1 = fig.add_subplot(121,title="Relion performance")
# ax1.set_ylabel("Iterations per minute")
# ax1.set_xlabel("GPU count")

ax2 = fig.add_subplot(111,title="Tensorflow performance")

titles  = ['']

for index,guy in enumerate(guys):
    titles.append(guy[1])
    ax2.barh(index,guy[0],color=guy[2])



# ax2.barh(np.arange(len(guys)),heights, color=colours, hatch='/')

ax2.set_xlabel("images/sec")

ax2.set_yticks(np.arange(-1,len(guys)+1))
ax2.set_yticklabels(titles)
# ax1.legend()
# ax1.set_axisbelow(True)
# ax1.yaxis.grid(color='lightgray')
# ax1.xaxis.grid(color='lightgray')
ax2.legend()
ax2.set_axisbelow(True)
ax2.yaxis.grid(color='lightgray')
ax2.xaxis.grid(color='lightgray')

plt.tight_layout()
fig.show()
```
